# T-Test Assignment

**ID:** 2215151011

### Given Data
- Data: [5.2, 5.4, 5.1, 5.6, 5.3, 5.8]
- Hypothesized Mean: 5.5

### Tasks:
1. Calculate the t-statistic and p-value.
2. Interpret and visualize the result at a 0.05 significance level.


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data and hypothesized mean
data = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])
hypothesized_mean = 5.5

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(data, hypothesized_mean)

# Print results
print(f"T-Statistic: {t_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (significant difference from hypothesized mean).")
else:
    print("Fail to reject the null hypothesis (no significant difference from hypothesized mean).")
```

```python
# Visualization
plt.figure(figsize=(8, 5))
plt.hist(data, bins=5, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(hypothesized_mean, color='red', linestyle='dashed', linewidth=2, label='Hypothesized Mean (5.5)')
plt.axvline(np.mean(data), color='green', linestyle='dashed', linewidth=2, label=f'Sample Mean ({np.mean(data):.2f})')
plt.legend()
plt.title("Data Distribution with Hypothesized Mean")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
